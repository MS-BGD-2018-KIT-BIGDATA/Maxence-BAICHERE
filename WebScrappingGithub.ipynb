```python
#import the library used to query a website
import urllib
import urllib.request
import requests
import json

#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
#import pandas to convert list to data frame
import pandas as pd

def getSoupFromURL(url):
    #Query the website and return the html to the variable 'page'
    page = urllib.request.urlopen(url)

    #Parse the html in the 'page' variable, and store it in Beautiful Soup format
    soup = BeautifulSoup(page, "lxml")
    return soup

def getWantedData(soup):
    A = []
    all_cells=soup.find('tbody')
    names = all_cells.find_all('tr')
    for name in names:
        pseudo = name.find('a')
        if pseudo != None:
            A.append(pseudo.text)
    return A

def extractDataFromWebPage(url):
    #specify the url
    url = "https://gist.github.com/paulmillr/2657075"
    soup = getSoupFromURL(url)
    data = getWantedData(soup)
    #df = convertDataIntoDF(data)
    return data

def convertDataIntoDF(A,B):    
    df=pd.DataFrame(A,columns = ['PSEUDO'])
    df["Average stars"] = B
    return df

myList = extractDataFromWebPage("https://gist.github.com/paulmillr/2657075")
```

```python
import time
def getAverageStars(userName, auth):
    repositories = requests.get("https://api.github.com/users/"+userName+"/repos", auth = auth).json()
    nbRepositories = 0
    nbStars = 0
    res = 0
    for repo in repositories:
        nbStars = nbStars + repo['stargazers_count']
        nbRepositories += 1
    if nbRepositories != 0:
        res = nbStars/nbRepositories
    return res
   
def getListeAverageStars(liste, auth):
    B = []  
    for element in liste:
        avg = getAverageStars(element, auth)
        B.append(avg)
    return B


def starsDataFrame(liste, auth):
    return convertDataIntoDF(liste, getListeAverageStars(liste, auth))

def getcontribsortedwithtime(auth):
    start_time = time.time()
    myList = extractDataFromWebPage("https://gist.github.com/paulmillr/2657075")
    result = starsDataFrame(myList, auth).sort_values('Average stars', ascending = False)
    t=(time.time() - start_time)
    m=int(t//60)
    s=round(t-m*60)
    minutes=str(m)
    secondes=str(s)
    return (result, minutes + ' minutes et ' + secondes +' secondes')

# Please enter your Github 'username' and 'password'
authperso=('', '')
getcontribsortedwithtime(authperso)                  
```
