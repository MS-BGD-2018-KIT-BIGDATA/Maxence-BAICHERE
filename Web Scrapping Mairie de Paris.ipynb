```python
#import the library used to query a website
import urllib
import urllib.request
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
#import pandas to convert list to data frame
import pandas as pd

#wiki = "http://alize2.finances.gouv.fr/communes/eneuro/detail.php?icom=056&dep=075&type=BPS&param=5&exercice=2013"

#Cell Data we are interested in
columnNames = ["TOTAL DES PRODUITS DE FONCTIONNEMENT = A", "TOTAL DES CHARGES DE FONCTIONNEMENT = B", "TOTAL DES RESSOURCES D'INVESTISSEMENT = C", "TOTAL DES EMPLOIS D'INVESTISSEMENT = D"]


def getSoupFromURL(url):
    #Query the website and return the html to the variable 'page'
    page = urllib.request.urlopen(url)

    #Parse the html in the 'page' variable, and store it in Beautiful Soup format
    soup = BeautifulSoup(page)
    return soup

def getWantedData(soup):
    A = []
    B = []
    C = []
    D = []
    for paragraph in soup.find_all(class_='bleu'):   
        cells = paragraph.findAll('td', class_='libellepetit')
        montant=paragraph.findAll('td', class_='montantpetit')
        if(len(montant)>0 and len(cells)>0):
            #print(montant[0].find(text=True))
            #A.append(montant[0].find(text=True))
            #C.append(montant[1].find(text=True))
            #D.append(montant[2].find(text=True))
            #print(cells[0].find(text=True))
            #B.append(cells[0].find(text=True))
            if(cells[0].find(text = True) in columnNames):
                #print(cells[0].find(text = True))
                A.append(montant[0].find(text=True))
                C.append(montant[1].find(text=True))
                D.append(montant[2].find(text=True))
                B.append(cells[0].find(text=True))
    return [A,B,C,D]
                

def convertDataIntoDF(liste):
    A = liste[0]
    B = liste[1]
    C = liste[2]
    D = liste[3]
    df=pd.DataFrame(B,columns = ['OPERATIONS'])
    df["En milliers d'Euros"] = A
    df['Euros par habitant'] = C
    df['Moyenne de la strate'] = D
    return df

def extractDataFromWebPage(year):
    #specify the url
    years = str(year)
    url = "http://alize2.finances.gouv.fr/communes/eneuro/detail.php?icom=056&dep=075&type=BPS&param=5&exercice="+years
    soup = getSoupFromURL(url)
    data = getWantedData(soup)
    df = convertDataIntoDF(data)
    return df

def concatenateData(yearBegin, yearEnd):
    df = extractDataFromWebPage(yearBegin)
    df['Year'] = yearBegin
    if yearEnd > yearBegin:
        for annee in range(yearBegin + 1, yearEnd+1):
            new_df = extractDataFromWebPage(annee)
            new_df['Year'] = annee
            df = pd.concat([df, new_df])
    return df

#Choose the range of the years you want to see the data
concatenateData(2010,2015)

```
